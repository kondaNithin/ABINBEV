```python
import pandas as pd

actuals_df = pd.read_csv('actuals.csv')
target_df = pd.read_csv('Targets.csv')
```

```python
group_columns = ['Plant','Year','Period']
sum_result = actuals_df.groupby(group_columns)['Quantity'].sum().reset_index()
```

```python
sum_result
```

```python
colfil ='Period'
Val = '7'
filtered = sum_result[sum_result[colfil] == 7]
```

```python
filtered
```

```python
group_columns = ['Plant','Year','Period']
sum_result_t = target_df.groupby(group_columns)['Target Quantity'].sum().reset_index()
```

```python
sum_result_t
```

```python
all = pd.merge(filtered,sum_result_t,on = "Plant",how="inner")
```

```python
all
```

```python
col1 = 'Quantity'
col2 = 'Target Quantity'
all['Variance in Quantity'] = all[col1]-all[col2]
```
